```python
import pandas as pd
#展示左右列
pd.set_option('display.max_columns', None)
```

```python
#读取application_data.csv
df = pd.read_csv('application_data.csv')
df
```

```python
#df的CODE_GENDER列变成0,1变量
df['CODE_GENDER']=[1 if x=='M' else 0 for x in df['CODE_GENDER']]
df['FLAG_OWN_CAR']=[1 if x=='Y' else 0 for x in df['FLAG_OWN_CAR']]
df['FLAG_OWN_REALTY']=[1 if x=='Y' else 0 for x in df['FLAG_OWN_REALTY']]
df
```

```python
df['NAME_EDUCATION_TYPE'].value_counts()
```

```python
education_mapping = {"Secondary / secondary special": 1,"Higher education": 2,"Incomplete higher": 3,"Lower secondary": 4,"Academic degree": 5}
df["NAME_EDUCATION_TYPE"] = df["NAME_EDUCATION_TYPE"].replace(education_mapping)
df
```

```python
df['NAME_INCOME_TYPE'].value_counts()
```

```python
#根据是否有工作，将NAME_INCOME_TYPE变成0,1变量
df['NAME_INCOME_TYPE']=[1 if x in ['Working','Commercial associate','State servant','Businessman'] else 0 for x in df['NAME_INCOME_TYPE']]
```

```python
#选取第三列到最后一列的数据
df = df.iloc[:,2:]
df
```

```python
#选取CODE_GENDER,FLAG_OWN_CAR,FLAG_OWN_REALTY,CNT_CHILDREN,AMT_INCOME_TOTAL,AMT_CREDIT,NAME_INCOME_TYPE,NAME_EDUCATION_TYPE列
#选取0-7列和40-53列
df = df.iloc[:,[0,1,2,3,4,5,6,7,38,57]]
```

```python
#所有列做Max-Min归一化除了TARGET列
for i in range(len(df.columns)-1):
    df.iloc[:,i] = (df.iloc[:,i]-df.iloc[:,i].min())/(df.iloc[:,i].max()-df.iloc[:,i].min())
```

```python
df.head()
```

```python
#df中所有数保留小数点后4位，并乘以10000，再取整，不包含最后一列
for i in range(len(df.columns)-1):
    df.iloc[:,i] = df.iloc[:,i].apply(lambda x: int(x*10000))
df.head()
```

```python
#导出为csv文件
df.to_csv('application_data_preprocessed.csv',index=False)
```

```python
#用KNN方法做分类
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
#选取前面所有列作为X
X = df.iloc[:,:-1]
#选取最后一列作为Y
Y = df.iloc[:,-1]
#划分训练集和测试集
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=0)
#KNN分类器
clf = KNeighborsClassifier(n_neighbors=3)
#训练
clf.fit(X_train,Y_train)
#预测
Y_pred = clf.predict(X_test)
#计算准确率
accuracy_score(Y_test,Y_pred)


```
